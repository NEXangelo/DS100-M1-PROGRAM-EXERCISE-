In this M1 programming exercise, we will utilize the dataset titled Annual plastic waste by disposal method, 2000 to 2019.
This dataset offers insights into the management of plastic waste over a span of two decades, providing valuable information on the methods employed for disposing of plastic waste.

DATA SOURCE: https://ourworldindata.org/grapher/plastic-fate?tab=table

```python
import requests as rq
from io import BytesIO
```

```python
import numpy as np
import pandas as pd
```

```python
url = 'https://github.com/NEXangelo/DS100-M1-PROGRAM-EXERCISE-/raw/main/DS-DATASET-plastic-fate.xlsx'
dt = rq.get(url).content
data = pd.read_excel(BytesIO(dt))
pd.set_option('display.max_rows',None)
data.head()
```

```python
#selecting the Countries USA and China only
country_data = data.loc[data['Entity'].str.contains('United States|China$')]
country_data.update(country_data[['Year','Recycled', 'Incinerated','Littered and mismanaged', 'Landfilled' ]].fillna(0))
if 'Code' in country_data.columns:
    country_data.drop(columns=['Code'], inplace=True)
country_data
```

```python
# Filter the DataFrame to include only the years from 2010 to 2019
country_data = country_data[(country_data['Year'] >= 2010) & (country_data['Year'] <= 2019)]
country_data
```

```python
country_data = country_data.copy()

# Calculate total trash and percentages
country_data['Total Trash (Tones)'] = (country_data['Recycled'] + country_data['Incinerated'] + country_data['Littered and mismanaged'] + country_data['Landfilled'])

# Calculate percentages
country_data['Recycled Percentage'] = np.round((country_data['Recycled'] / country_data['Total Trash (Tones)']) * 100, 2)
country_data['Incinerated Percentage'] = np.round((country_data['Incinerated'] / country_data['Total Trash (Tones)']) * 100, 2)
country_data['Littered and mismanaged Percentage'] = np.round((country_data['Littered and mismanaged'] / country_data['Total Trash (Tones)']) * 100, 2)
country_data['Landfilled Percentage'] = np.round((country_data['Landfilled'] / country_data['Total Trash (Tones)']) * 100, 2)

# Fill NaN values with 0
country_data.update(country_data[['Recycled Percentage', 'Incinerated Percentage', 'Littered and mismanaged Percentage', 'Landfilled Percentage']].fillna(0))
country_data = country_data[['Entity', 'Year', 'Total Trash (Tones)', 'Recycled Percentage', 'Incinerated Percentage', 'Littered and mismanaged Percentage', 'Landfilled Percentage']]

# Display the updated DataFrame
country_data
```

```python
# Calculate 'Sustainable Waste Management' percentage
country_data['Sustainable Waste Management Percentage'] = (country_data['Recycled Percentage'] + country_data['Incinerated Percentage'])

# Format the percentages with a percentage sign
country_data['Sustainable Waste Management Percentage'] = country_data['Sustainable Waste Management Percentage']

# Calculate 'Mismanaged Waste Disposal' percentage
country_data['Mismanaged Waste Disposal Percentage'] = (country_data['Littered and mismanaged Percentage'] + country_data['Landfilled Percentage'])

# Format the percentages with a percentage sign
country_data['Mismanaged Waste Disposal Percentage'] = country_data['Mismanaged Waste Disposal Percentage']

# Display the updated DataFrame with required columns
country_data[['Entity', 'Year', 'Sustainable Waste Management Percentage', 'Mismanaged Waste Disposal Percentage']]
```

```python
# Calculate the ratio of Sustainable Waste Management Percentage to Mismanaged Waste Disposal Percentage for each entity
country_data['Sustainable_to_Mismanaged_Ratio'] = country_data.groupby('Entity')['Sustainable Waste Management Percentage'].transform(lambda x: x / country_data['Mismanaged Waste Disposal Percentage'])

# Compute the difference between consecutive years' ratios for each entity
country_data['Trend (Yearly Change in SWM:MWD Ratio)'] = country_data.groupby('Entity')['Sustainable_to_Mismanaged_Ratio'].diff()

# Fill NaN values in the trend column with 0
country_data['Trend (Yearly Change in SWM:MWD Ratio)'] = country_data['Trend (Yearly Change in SWM:MWD Ratio)'].fillna(0)

# Display the updated DataFrame with required columns
required_columns = ['Entity', 'Year', 'Sustainable Waste Management Percentage', 'Mismanaged Waste Disposal Percentage', 'Trend (Yearly Change in SWM:MWD Ratio)']
country_data[required_columns]
```

```python
# Filter data for China and the United States
china_data = country_data[country_data['Entity'] == 'China']
usa_data = country_data[country_data['Entity'] == 'United States']

# Merge the data frames on the 'Year' column
merged_data = pd.merge(china_data, usa_data, on='Year', suffixes=('_China', '_USA'))

# Format columns as percentages
merged_data['Sustainable Waste Management Percentage_China'] = merged_data['Sustainable Waste Management Percentage_China'].map('{:.2f}%'.format)
merged_data['Mismanaged Waste Disposal Percentage_China'] = merged_data['Mismanaged Waste Disposal Percentage_China'].map('{:.2f}%'.format)
merged_data['Sustainable Waste Management Percentage_USA'] = merged_data['Sustainable Waste Management Percentage_USA'].map('{:.2f}%'.format)
merged_data['Mismanaged Waste Disposal Percentage_USA'] = merged_data['Mismanaged Waste Disposal Percentage_USA'].map('{:.2f}%'.format)

# Display the merged data frame with required columns
required_columns = ['Year', 
                    'Entity_China', 'Sustainable Waste Management Percentage_China', 'Mismanaged Waste Disposal Percentage_China','Trend (Yearly Change in SWM:MWD Ratio)_China',
                    'Entity_USA', 'Sustainable Waste Management Percentage_USA', 'Mismanaged Waste Disposal Percentage_USA','Trend (Yearly Change in SWM:MWD Ratio)_USA']

merged_data = merged_data[required_columns]
merged_data



```

```python
# Rename the columns for better presentation
merged_data = merged_data.rename(columns={
    'Entity_China': 'Country (China)',
    'Sustainable Waste Management Percentage_China': 'Sustainable Waste Management Percentage (China)',
    'Mismanaged Waste Disposal Percentage_China': 'Mismanaged Waste Disposal Percentage (China)',
    'Entity_USA': 'Country (USA)',
    'Sustainable Waste Management Percentage_USA': 'Sustainable Waste Management Percentage (USA)',
    'Mismanaged Waste Disposal Percentage_USA': 'Mismanaged Waste Disposal Percentage (USA)',
    'Trend (Yearly Change in SWM:MWD Ratio)_China': 'Trend (Yearly Change in SWM:MWD Ratio) (China)',
    'Trend (Yearly Change in SWM:MWD Ratio)_USA': 'Trend (Yearly Change in SWM:MWD Ratio) (USA)'
})

# Display the updated data frame with the new column names
merged_data
```

The decision to focus on China and the United States was based on their global influence and their differing approaches to environmental management.

To process the data, we first filtered the dataset to include only information for the United States and China. Then, we computed the total waste generated by summing up the values for recycling, incineration, littering, mismanaged waste, and landfilling. This allowed us to translate the measured weight of trash into percentages, representing the proportion of each waste management method relative to the total plastic waste generated.

Next, we categorized the waste into two main categories: sustainable waste management and mismanaged waste disposal. Sustainable waste management includes recycling and incineration, while mismanaged waste disposal encompasses littering and landfilling. By computing the percentages for these two categories, we gained insights into the effectiveness of waste management practices in each country.

Finally, we computed the trend in the yearly change of the ratio between sustainable waste management and mismanaged waste disposal percentages. This trend analysis helps identify patterns and variations in waste management practices over time, providing valuable information for decision-making and policy formulation aimed at achieving sustainable waste management goals.
